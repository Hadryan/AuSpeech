```python
import warnings
warnings.filterwarnings("ignore")

from run_generation_continuous import initiate_model, get_response, MODEL_CLASSES
import pandas as pd
import datetime
import nltk
```

```python
df = pd.read_csv("train_sent_emo.csv")
```

```python
df.head()
```

```python
rows = df.to_dict('records')
```

```python
rows[0]
```

```python
rows[0]['Utterance'].encode('ascii', 'ignore').decode('utf8').lower()
```

```python
conversations = []
for idx in range(len(rows)):
    if idx + 1 < len(rows) and rows[idx]['Dialogue_ID'] == rows[idx + 1]['Dialogue_ID']:
        conv = {
            "Utterance": rows[idx]["Utterance"].encode('ascii', 'ignore').decode('utf8').lower(),
            "Reply": rows[idx + 1]["Utterance"].encode('ascii', 'ignore').decode('utf8').lower(),
            "Season": rows[idx]["Season"],
            "Episode": rows[idx]["Episode"],
            "Dialogue_ID": rows[idx]["Dialogue_ID"]
        }
        conversations.append(conv)
```

```python
conv_df = pd.DataFrame(conversations)
```

```python
conv_df = pd.read_csv("formatted_replies.csv")
```

```python
conv_df.head()
```

```python
model, tokenizer, args = initiate_model()
```

```python
counter = 0
def get_generated_text(utterance):
    global counter
    counter += 1
    response = get_response(model, tokenizer, utterance, args)
    response = response.replace(utterance, "").replace("\n", "").strip()
    print("{} Generating {}/{} {}".format(datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"), 
                                          counter, len(rows), response))
    return response
```

```python
get_response(model, tokenizer, "Hello", args)
```

```python
conv_df["Generated_Reply"] = conv_df["Utterance"].apply(get_generated_text)
```

```python
conv_df.to_csv("formatted_replies.csv", index=False)
```

```python
corpus = [row["Utterance"].encode('ascii', 'ignore').decode('utf8').split() for row in rows]
```

```python
counter = 0
def calculate_bleu(reply):
    global counter
    bleu = nltk.translate.bleu_score.sentence_bleu(corpus, reply)
    print("{} Calculating Bleu {}/{} - {}".format(datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"), 
                                          counter, len(rows), bleu))
    counter += 1
    return bleu
```

```python
conv_df["bleu"] = conv_df["Generated_Reply"].apply(calculate_bleu)
```
